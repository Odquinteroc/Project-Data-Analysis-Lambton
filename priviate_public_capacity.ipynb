```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("annual_capacity_Privete_Public .csv")

df
```

```python
df = df.drop(columns=['DGUID', 'UOM_ID', 'SCALAR_FACTOR', 'SCALAR_ID', 'VECTOR', 'COORDINATE', 'STATUS', 'SYMBOL', 'TERMINATED', 'DECIMALS'])
df
```

```python
df.isna().sum()
```

```python
df['VALUE'] = df['VALUE'].fillna((df['VALUE'].shift() + df['VALUE'].shift(-1)) / 2)
df
```

```python
tipos_diferentes = df['Class of electricity producer'].unique()
dataframes_separados = {}
for tipo in tipos_diferentes:
    # Seleccionar las filas que tienen el tipo actual
    df_tipo = df[df['Class of electricity producer'] == tipo].copy()  # Usar copy para evitar la vista
    # Agregar el DataFrame al diccionario con el tipo como clave
    dataframes_separados[tipo] = df_tipo.reset_index(drop=True)
```

```python
for k, v in dataframes_separados.items():
    print(f"DataFrame para tipo '{k}':")
    print(v)
```

```python
for k, v in dataframes_separados.items():
    # Verificar si hay al menos una fila en el DataFrame
    if len(v) > 0:
        primer_valor = v.loc[v.index[0], 'VALUE']  # Acceder al primer valor de 'VALUE'
        v['Porcentaje_Aumento'] = ((v['VALUE'] - primer_valor) / primer_valor) * 100
        print(f"DataFrame para tipo '{k}':")
        print(v)
    else:
        print(f"No hay datos en el DataFrame para tipo '{k}'")
```

```python
import os

for k, v in dataframes_separados.items():
    v.plot( x ='REF_DATE', y = 'VALUE', kind='line', figsize=(10, 6))
    plt.title(f' {k}')
    plt.xlabel('Year')
    plt.ylabel('Energy Generation')
    plt.xticks( rotation=45)
    plt.grid(True)
    plt.ylim(min(v['VALUE']) - 10, max(v['VALUE']) + 10)
    # Show the plot
    plt.legend(title='Type of Energy', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    v.to_csv(os.path.join("private_public_capacity/", f"capacity in {k}.csv"))
    plt.savefig(os.path.join("private_public_capacity/", f"Generation of {k}"), bbox_inches = "tight", dpi=300)
```

```python
for k, v in dataframes_separados.items():
    v.plot( x ='REF_DATE', y = 'Porcentaje_Aumento', kind='line', figsize=(10, 6))
    plt.title(f' {k}')
    plt.xlabel('Year')
    plt.ylabel('Energy Generation')
    plt.xticks( rotation=45)
    plt.grid(True)
    plt.ylim(min(v['Porcentaje_Aumento']) - 10, max(v['Porcentaje_Aumento']) + 10)
    # Show the plot
    plt.legend(title='Type of Energy', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.savefig(os.path.join("private_public_capacity/", f"increase of generation in %  of{k}"), bbox_inches = "tight", dpi=300)
```
